```python
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns = iris_data.feature_names)
iris['target'] = iris_data.target
iris['species'] = iris['target'].map(dict(enumerate(iris_data.target_names)))
iris.pop('target')
iris = iris.rename(columns={
    "sepal length (cm)" : "sepal_length",
    "sepal width (cm)" : "sepal_width",
    "petal length (cm)" : "petal_length",
    "petal width (cm)" : "petal_width"
})
iris
```

```python
iris.to_csv("iris.csv", index = False)
```

EX 1

leia koguhind
kui hind üle 100, siis 20% allahindlus
kui hind üle 50, siis 10% allahindlus

```python
hinnad = [120, 45, 89, 200, 33, 75, 150, 10, 99, 101]

koguhind = sum(hinnad) 
print(f"Koguhind on: {koguhind}")

for hind in hinnad:
    if hind > 100:
        print(f"Hind: {hind}, allahindlus: 20%, allahinnatud hind: {round (hind*0.8)}")
    elif hind > 50:
        print(f"Hind: {hind}, allahindlus: 10%, allahinnatud hind: {round (hind*0.9)}")
```

```python
def get_discount(price: int) -> float:
    if price > 100:
        return price * 0.8
    if price > 50:
        return price * 0.9
    return price

print(get_discount(101))
print(get_discount(72))
print(get_discount(8))
```

```python
hinnad = [120, 45, 89, 200, 33, 75, 150, 10, 99, 101]

get_price(hinnad)

def get_price(suva_nimi: list[int]) -> float:
    tulemus = 0
    for hind in suva_nimi:
        allahindlusega_hind = get_discount (hind)
        tulemus = tulemus + allahindlusega_hind
    return tulemus

def get_price_short(suva_nimi: list[int]) -> float:
    return sum([get_discount(hind) for hind in suva_nimi])
get_price_short(hinnad)
```
